```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {}
```

```python
print("- Setup inventory -")
for item in products:
    quantity= int(input(f"Add the inventory for {item} items, please: "))
    inventory[item] = quantity
```

```python
customer_orders = set()
print("\n- Place Your Orders -)")
for i in range (3):
    choice = input(f"Write the name of the product {i+1}: ")
    customer_orders.add(choice)
```

```python
total_inventory_items = len(products)
total_orders = len(customer_orders)
percentage = (total_orders / total_inventory_items) * 100
order_status = (total_orders, percentage)
```

```python
print("\n--- Order Statistics ---")
print(f"Total Products Ordered: {order_status[0]}")
print(f"Percentage of Products Ordered: {order_status[1]:.2f}%")
```

```python
print("\n--- Updating Inventory ---")
for product in customer_orders:
    if product in inventory:
        inventory[product] -= 1
        print(f"Inventory updated: {product} now has {inventory[product]} units.")
    else:
        print(f"Product '{product}' not found in original inventory.")
```

```python
print("\n--- Final Inventory Status ---")
for product, quantity in inventory.items():
    print(f"Product: {product:10} | Remaining: {quantity}")
```
